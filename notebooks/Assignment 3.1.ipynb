### import packages

```python
import pandas as pd
import numpy as np
```

### Load train and test data

```python
train_data_path = r"../data/processed/train.csv"
test_data_path = r"../data/processed/test.csv"
```

```python
train_data = pd.read_csv(train_data_path)
test_data  = pd.read_csv(test_data_path)
print(f"train data shape : {train_data.shape}")
print(f"test data shape : {test_data.shape}")
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
test_data.rename(columns={'Unnamed: 0':'index'}, inplace=True)
```

```python
test_data.set_index("index")
```

```python
train_data.rename(columns={'Unnamed: 0':'index'}, inplace=True)
train_data.set_index("index")
```

```python
housing_prepared = train_data.drop(['median_house_value'],axis=1)
housing_labels = train_data['median_house_value']
```

### Hyper Parameter Tuning

```python
from sklearn.svm import SVR
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import expon, reciprocal, uniform
```

```python
param_distribs = {
        'kernel': ['linear', 'rbf'],
        'C': uniform(1, 10),
        'gamma': expon(scale=1.0),
    }

svm_reg = SVR()
rnd_search = RandomizedSearchCV(svm_reg, param_distributions=param_distribs,n_iter =5,
                                cv=5, scoring='neg_root_mean_squared_error',
                                verbose=2, random_state=42)
rnd_search.fit(housing_prepared, housing_labels)
```

```python
rnd_search.best_params_
```

```python
cvres = rnd_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

### Evaluation

```python
from sklearn.metrics import mean_squared_error
```

```python
final_model = rnd_search.best_estimator_
```

```python
test_prepared = test_data.drop(['median_house_value'],axis=1)
test_labels = test_data['median_house_value']
```

```python
final_predictions = final_model.predict(test_prepared)

final_mse = mean_squared_error(final_predictions, test_labels)
final_rmse = np.sqrt(final_mse)
print("RMSE : ", final_rmse)
```
